````python
# Teste de consulta e devolução dos dados em DataFrame
uf = "MG"
tipo_interessado = "Estado"
base_url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
parametros = {'uf': uf, 'tipo_interessado': tipo_interessado}
resposta = requests.get(base_url, params=parametros)
data_json = resposta.json()
````

````python
#função com o resultado

def consultar_dados(uf, tipo_interessado):
    base_url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
    parametros = {'uf': uf, 'tipo_interessado': tipo_interessado}
    
    try:
        resposta = requests.get(base_url, params=parametros)
        if resposta.status_code == 200:
            data_json = resposta.json()
            df = pd.DataFrame(data_json['items'])
            return df
        else:
            print(f"Erro na solicitação HTTP (Código de Status {resposta.status_code}).")
            return None
    except Exception as e:
        print(f"Erro ao consultar dados: {str(e)}")
        return None

    
uf = "MG"
tipo_interessado = "Estado"
df_resultado = consultar_dados(uf, tipo_interessado)
````

````python
# 2) Seu código aqui
status_arquivadas = (df_resultado['status'] == 'Arquivado por decurso de prazo')
qtd_status_arquivadas = len(status_arquivadas)
print(f"A quantidade de solicitações para o Estado de Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas são de {qtd_status_arquivadas}") 
````

````python
# 3) Seu código aqui
uf = "BA"
tipo_interessado = "Município"
df_resultado = consultar_dados(uf, tipo_interessado)

top_deferidos = df_resultado[df_resultado['status'] == 'Deferido']['interessado'].value_counts()
print(top_deferidos.head(1))
````

````python
# 4) Seu código aqui
uf = "BA"
tipo_interessado = "Estado"
df_resultado = consultar_dados(uf, tipo_interessado)
df_resultado.to_csv('solicitacoes_bahia_estado.csv')
````
